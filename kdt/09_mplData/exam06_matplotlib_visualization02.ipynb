```python
import math

def sigmoid(x):
    sig = 1 / (1+math.exp(-x))
    return sig

point = 100
X = np.linspace(-10, 10, point)
Y = list(map(sigmoid, X))

fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0, 0, 1, 1])

np.random.seed(100)
N = 50

ax.plot(X, Y, color='r', linewidth=2)

ax.fill_between(X, list(map(lambda Y: Y+0.25, Y)),
                list(map(lambda Y: Y-0.25, Y)),
                facecolor='r', edgecolor='None', alpha=0.15)

plt.axhline(color='black', linewidth=1)
plt.axvline(c='k', linewidth=1)
ax.set_title('Sigmoid')
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Hogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 0.1만큼 혼자 튀어나옴
fig1, ax1 = plt.subplots()

wedges, texts, autotexts = ax1.pie(
    sizes, labels=labels, explode=explode, shadow=True, startangle=-90, autopct='%.2f%%')  # %를 넣고싶으면 %%
ax1.axis('equal')
wedges[0].set(hatch='///', alpha=0.9)
wedges[2].set_radius(1.1)  # 반지름을 키워서 강조
plt.show()
```

```python
size = 0.3
vals1 = [40, 35, 25]
vals2 = [11, 29, 12, 23, 13]

cmap = plt.get_cmap('tab20c')
outer_colors = cmap([0, 4, 8])
inner_colors = cmap([1, 2, 5, 6, 9, 10])

fig = plt.figure(figsize=(8, 5))
ax = fig.add_axes([0, 0, 1, 1])
ax.pie(vals1, radius=1, colors=outer_colors,
       labels=['male', 'female', 'pet'],
       autopct='%1.1f%%', pctdistance=0.83,
       textprops={'color': 'black'},
       wedgeprops={'width': size, 'edgecolor': 'white', 'linewidth': 5},
       startangle=0,
       counterclock=True)
# pctdistance 퍼센트위치 지정해주는 거

ax.pie(vals2, radius=1, colors=inner_colors,
       autopct='%1.1f%%', pctdistance=0.8,
       textprops={'color': 'black'},
       wedgeprops={'width': size, 'edgecolor': 'white', 'linewidth': 5},
       startangle=0,
       counterclock=True)
plt.show()
```
