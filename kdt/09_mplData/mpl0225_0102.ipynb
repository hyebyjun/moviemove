```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False     # 이거 안해주면 마이너스 깨짐..
```

```python
x = [1, 2, 3]
y = [2, 4, 8]

plt.plot(x, y)   # 선
plt.bar(x, y)    # 막대
plt.title('wahoooo')
```
